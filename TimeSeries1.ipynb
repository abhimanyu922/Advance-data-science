# **Time Series Analysis - GDP Data**

```python
# Step 0 - Import Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1 - Data Loading
df = pd.read_csv('/content/India_GDP_1960-2022.csv',skiprows=1)
df.head()
```

```python
# 1.1 Data Investigation - Information
df.info()
```

```python
# drop the unnamed column
df.drop('Unnamed: 0',axis=1,inplace=True)
df.head()
```

```python
# Step 2 - Data Cleaning
# Step 2.1 - Missing Data
df.isnull().sum()
```

```python
# Step 2.2 - Duplicated Data
df.duplicated().sum()
```

```python
# Step 2.3  - Outliers
df.describe()
```

```python
# Visualize the result
plt.figure(figsize=(10,5))

for i,col in enumerate(df.columns):
  plt.subplot(2,2,i+1)
  plt.title(col)
  sns.boxplot(df[col])
  plt.show()
```

```python
# Step 2.4 - Transformation
# Change the year column into datetime
df['Year'] = pd.to_datetime(df['Year'],format='%Y')
df.head()
```

```python
# Step 3 - EDA
# Trend Analysis
sns.lineplot(x='Year',y='GDP in (Billion) $',data=df)
```

```python
sns.lineplot(x='Year',y='Growth %',data=df)
```
