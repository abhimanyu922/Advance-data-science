Python example: Creating and visualizing a simple dataset

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a sample dataset
heights = np.random.normal(170, 10, 1000)  # 1000 heights with mean 170cm and std dev 10cm

plt.figure(figsize=(10, 6))
plt.hist(heights, bins=30, edgecolor='black')
plt.axvline(np.mean(heights), color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(np.median(heights), color='orange', linestyle='dashed', linewidth=2, label='Median')
plt.legend()
plt.title('Distribution of Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.show()
```

## 2.1 Measures of Central Tendency

The three main measures of central tendency are:
1. Mean
2. Median
3. Mode

![](https://miro.medium.com/v2/resize:fit:1204/1*6dzEJhNiHLfKG385hgtSvg.jpeg)

Let's use Seaborn's 'tips' dataset to illustrate these concepts.

```python
# tips dataset from seabron
import seaborn as sns
tips = sns.load_dataset("tips")
tips.head()
```

```python
# Calculate the central tendency of total_bill
mean_total_bill = tips['total_bill'].mean()
median_total_bill = tips['total_bill'].median()
mode_total_bill = tips['total_bill'].mode()

print(f"Mean: {mean_total_bill}")
print(f"Median: {median_total_bill}")
print(f"Mode: {mode_total_bill}")
```

```python
# Visualize the distribution
plt.figure(figsize=(9, 4))
sns.histplot(tips['total_bill'], bins=20, kde=True)
plt.axvline(mean_total_bill, color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(median_total_bill, color='orange', linestyle='dashed', linewidth=2, label='Median')
plt.legend()
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')
plt.show()
```

## 2.2 Measures of Variability

Measures of variability describe how spread out the data is. Common measures include:
1. Range
2. Variance
3. Standard Deviation

Let's use the 'penguins' dataset to illustrate these concepts.

```python
# Load the penguins dataset
penguins = sns.load_dataset("penguins")

# Calculate range, variance, and standard deviation of flipper length
flipper_range = penguins['flipper_length_mm'].max() - penguins['flipper_length_mm'].min()
flipper_var = penguins['flipper_length_mm'].var()
flipper_std = penguins['flipper_length_mm'].std()

print(f"Flipper length range: {flipper_range:.2f} mm")
print(f"Flipper length variance: {flipper_var:.2f} mm")
print(f"Flipper length standard deviation: {flipper_std:.2f} mm")

# Visualize the distribution by species
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='flipper_length_mm', data=penguins)
plt.title('Flipper Length Distribution by Penguin Species')
plt.ylabel('Flipper Length (mm)')
plt.show()
```

## 2.3 Measures of Position

Measures of position help us understand where a particular value falls within a dataset. Common measures include:
1. Percentiles
2. Quartiles

Let's use the 'diamonds' dataset to explore these concepts.

```python
# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Calculate quartiles of diamond prices
q1 = diamonds['price'].quantile(0.25)
q2 = diamonds['price'].quantile(0.50)  # This is the median
q3 = diamonds['price'].quantile(0.75)

print(f"First quartile (Q1) of diamond prices: ${q1:.2f}")
print(f"Second quartile (Q2) of diamond prices: ${q2:.2f}")
print(f"Third quartile (Q3) of diamond prices: ${q3:.2f}")

# Visualize the distribution with quartiles
plt.figure(figsize=(12, 6))
sns.histplot(diamonds['price'], kde=True)
plt.axvline(q1, color='r', linestyle='--', label='Q1')
plt.axvline(q2, color='g', linestyle='--', label='Q2 (Median)')
plt.axvline(q3, color='b', linestyle='--', label='Q3')
plt.legend()
plt.title('Distribution of Diamond Prices with Quartiles')
plt.xlabel('Price ($)')
plt.show()
```

## 2.4 Measures of Relationship

When dealing with two or more variables, we often want to understand how they relate to each other. Common measures include:
1. Covariance
2. Correlation

Let's use the 'iris' dataset to explore these concepts.

```python
# Load the iris dataset
iris = sns.load_dataset("iris")

# Calculate correlation between sepal length and width
correlation = iris['sepal_length'].corr(iris['sepal_width'])

print(f"Correlation between sepal length and width: {correlation:.2f}")

# Visualize the relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title('Relationship between Sepal Length and Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

# 3. Probability

Probability is the study of randomness and uncertainty. It forms the foundation for statistical inference and is crucial for understanding many real-world phenomena.

We'll use the Titanic dataset to explore probability concepts. This dataset contains information about passengers on the Titanic, including whether they survived or not.

First, let's load and examine our dataset:

```python
# Load the Titanic dataset
titanic = sns.load_dataset("titanic")

print(titanic.head())
print("\nDataset Info:")
print(titanic.info())
```

## 3.1 Basic Probability Concepts

### Sample Space and Events

The sample space is the set of all possible outcomes of an experiment. An event is a subset of the sample space.

In our Titanic dataset:
- The sample space could be all passengers.
- An event could be "passenger survived" or "passenger was female".

Let's calculate some basic probabilities:

```python
total_passengers = len(titanic)
survivors = titanic['survived'].sum()
females = (titanic['sex'] == 'female').sum()

print(f"Total passengers: {total_passengers}")
print(f"Passengers who survived: {survivors}")
print(f"Passengers who were female: {females}")
```

```python
prob_survival = survivors / total_passengers
prob_female = females / total_passengers

print(f"Probability of survival: {prob_survival:.4f}")
print(f"Probability of being female: {prob_female:.4f}")

```

```python
# Visualize these probabilities
plt.figure(figsize=(10, 6))
sns.barplot(x=['Survival', 'Female'], y=[prob_survival, prob_female])
plt.title('Basic Probabilities in Titanic Dataset')
plt.ylabel('Probability')
plt.ylim(0, 1)
plt.show()
```

## 3.2 Conditional Probability

Conditional probability is the probability of an event occurring, given that another event has already occurred.

Let's calculate the probability of survival given that a passenger was female:

```python
female_survivors = titanic[(titanic['sex'] == 'female') & (titanic['survived'] == 1)].shape[0]
prob_survival_given_female = female_survivors / females

print(f"Probability of survival given female: {prob_survival_given_female:.4f}")

# Visualize conditional probabilities
survival_by_sex = titanic.groupby('sex')['survived'].mean()

plt.figure(figsize=(10, 6))
sns.barplot(x=survival_by_sex.index, y=survival_by_sex.values)
plt.title('Probability of Survival by Sex')
plt.ylabel('Probability of Survival')
plt.ylim(0, 1)
plt.show()
```

## 3.3 Independence

Two events are independent if the occurrence of one does not affect the probability of the other.

Let's check if survival was independent of passenger class:

```python
prob_first_class = (titanic['pclass'] == 1).mean()
prob_survival_first_class = titanic[titanic['pclass'] == 1]['survived'].mean()

print(f"P(Survival): {prob_survival:.4f}")
print(f"P(First Class): {prob_first_class:.4f}")
print(f"P(Survival | First Class): {prob_survival_first_class:.4f}")

```

```python
if abs(prob_survival - prob_survival_first_class) < 0.01:
    print("Survival and being in first class appear to be independent.")
else:
    print("Survival and being in first class do not appear to be independent.")

# Visualize survival probability by class
survival_by_class = titanic.groupby('pclass')['survived'].mean()

plt.figure(figsize=(10, 6))
sns.barplot(x=survival_by_class.index, y=survival_by_class.values)
plt.title('Probability of Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Probability of Survival')
plt.ylim(0, 1)
plt.show()
```

## 3.4 Bayes' Theorem

Bayes' Theorem is a way to update probabilities based on new evidence. It's expressed as:

**P(A|B) = (P(B|A) * P(A)) / P(B)**

Let's use Bayes' Theorem to calculate the probability that a passenger was in first class, given that they survived:

```python
P_survival = prob_survival
P_first_class = prob_first_class
P_survival_given_first_class = prob_survival_first_class

P_first_class_given_survival = (P_survival_given_first_class * P_first_class) / P_survival

print(f"P(First Class | Survival) = {P_first_class_given_survival:.4f}")

# Visualize the comparison
plt.figure(figsize=(10, 6))
sns.barplot(x=['P(First Class)', 'P(First Class | Survival)'],
            y=[P_first_class, P_first_class_given_survival])
plt.title('Probability of Being in First Class: Overall vs Given Survival')
plt.ylabel('Probability')
plt.ylim(0, 1)
plt.show()
```

## 3.5 Probability Distributions

A probability distribution describes the probability of each possible outcome in a sample space.

Let's look at the distribution of ages in our dataset:

```python
plt.figure(figsize=(12, 6))
sns.histplot(titanic['age'].dropna(), kde=True, bins=30)
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Calculate some probabilities from this distribution
prob_under_18 = (titanic['age'] < 18).mean()
prob_over_60 = (titanic['age'] > 60).mean()

print(f"Probability of being under 18: {prob_under_18:.4f}")
print(f"Probability of being over 60: {prob_over_60:.4f}")
```
